```python
### This code is to find the next valid time with given digits in HH:MM format. 
### For example; time 19:34 should return 19:39 (5 minutes later) 19:39 (has only digit from given time 1,9,3,4)
### 23:59 should return 22:22 (after 22 hours 23 minutes), 22:22 (has only digit used from 2,3,5,9)
from itertools import product

def next_valid_time(time):

    #Convert the times in minutes, which is easy to get the next time with minimum difference in minutes
    hours,minutes=time.split(":")
    total_current_minutes=int(hours)*60+int(minutes)
    print "current time in minutes", total_current_minutes
    
    
    all_digits=set(time.replace(':','')) #Get all the possible time from given digits
    print "all digits in the current time",all_digits
    
    
    valid_times={}
    all_combinations=list(product(all_digits,repeat=4)) #Get all possible permutations with given digits
    for i in all_combinations:
        hr=int("".join(i[0:2]))
        mn=int("".join(i[2:4]))
        print "Iterating over ", i, hr ,":", mn
        

        ### keep only valid times in possible combinations
        ### Find next minutes for each valid times

        if (hr < 24) & (mn < 60):
            next_min = hr*60 + mn - total_current_minutes
            if next_min <= 0:
                next_min += 24 * 60
                ##Move the time from previous times from today to tomorrow 
                valid_times[str('{:02}'.format(hr)) + ":" + str('{:02}'.format(mn))]=next_min
            else:
                next_min = hr*60 + mn - total_current_minutes
                valid_times[str('{:02}'.format(hr)) + ":" + str('{:02}'.format(mn))] = next_min
                
    print "All next valid times with minutes as values", valid_times
    next_closest_time=min(valid_times.keys(), key=(lambda k: valid_times[k])) 
    print "Next closest time with given digits is ", next_closest_time
    return next_closest_time
```

```python
next_valid_time("19:34")
```

```python
next_valid_time("23:59")
```

```python
next_valid_time("00:00")
```
